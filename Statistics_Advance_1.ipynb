1. What is a random variable in probability theory?
~ A random variable is a variable that takes on different values based on the outcome of a random event. It links outcomes to numbers.

2. What are the types of random variables?
~ There are two main types of random variables:
Discrete Random Variable – Takes countable values.
Example: Number on a die (1, 2, 3, ...)
Continuous Random Variable – Takes infinite values in a range.
Example: Height, weight, or temperature.

3. What is the difference between discrete and continuous distributions?
~ Discrete distribution deals with countable outcomes.
Example: Tossing a coin (Heads or Tails)
Continuous distribution deals with uncountable outcomes within a range.
Example: Measuring time or temperature (like 23.5°C, 23.51°C, etc.)

4. What are probability distribution functions (PDF)?
~ A Probability Distribution Function (PDF) shows how probabilities are distributed over values of a random variable.

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
~ CDF (Cumulative Distribution Function) gives the probability that a value is less than or equal to a certain number.
PDF (Probability Distribution Function) gives the probability of a specific value (discrete) or probability density (continuous).

6. What is a discrete uniform distribution?
~ A discrete uniform distribution is when all outcomes have equal probability.
Example: Rolling a fair die — each number (1 to 6) has a 1/6 chance.

7. What are the key properties of a Bernoulli distribution?
~ Bernoulli distribution has these key properties:
Two outcomes: Success (1) or Failure (0)
One trial only
Probability of success = p, failure = 1 − p
Mean = p
Variance = p(1 − p)

8. What is the binomial distribution, and how is it used in probability?
~ The binomial distribution gives the probability of getting a certain number of successes in a fixed number of independent trials, each with the same success chance.
Used for: Counting successes (like heads in 10 coin tosses)

9. What is the Poisson distribution and where is it applied?
~ The Poisson distribution models the number of times an event happens in a fixed time or space, when events occur independently and at a constant rate.
Used for: Counting rare events

10. What is a continuous uniform distribution?
~ A continuous uniform distribution means all values in a range are equally likely.
Example: Choosing a random number between 0 and 1 — every value like 0.2, 0.5, 0.99 has the same chance.

11. What are the characteristics of a normal distribution?
~ A normal distribution has these key characteristics:
Bell-shaped and symmetric
Mean = Median = Mode
Most values are close to the mean
Follows the 68-95-99.7 rule

12. What is the standard normal distribution, and why is it important?
~ The standard normal distribution is a normal distribution with:
Mean (μ) = 0
Standard deviation (σ) = 1
Importance:
Makes it easy to compare different normal distributions
Used to calculate Z-scores and find probabilities using standard tables

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
~ The Central Limit Theorem (CLT) says:
When you take many random samples from any population, the sample means will form a normal distribution as the sample size gets large (n ≥ 30), no matter the original population shape.
Why it's important:
Allows us to use normal distribution for inference, like confidence intervals and hypothesis testing—even if data isn't normally distributed.

14. How does the Central Limit Theorem relate to the normal distribution?
~ The Central Limit Theorem explains why the normal distribution appears so often in statistics.
It says that the distribution of sample means becomes normal (bell-shaped) as the sample size grows, even if the original data is not normal.
This links many real-world processes to the normal distribution.

15. What is the application of Z statistics in hypothesis testing?
~ Z-statistics is used in hypothesis testing to:
Compare sample data to a population
Check if the difference is significant when population standard deviation is known
Applications:
Testing means (e.g., is average height = 170 cm?)
Used in Z-test to calculate the Z-score, then compare it to critical values for decision-making.

16. How do you calculate a Z-score, and what does it represent?
~ A Z-score, also known as a standard score, measures how many standard deviations a data point is from the mean of a dataset. It's calculated by subtracting the mean from the data point and then dividing the result by the standard deviation.
What it represents:
A Z-score shows how many standard deviations a value is from the mean.
Z = 0 → at the mean
Z > 0 → above the mean
Z < 0 → below the mean

17. What are point estimates and interval estimates in statistics?
~ Point Estimate
A single value used to estimate a population parameter.
Interval Estimate
A range of values (with confidence level) likely to contain the population parameter.

18. What is the significance of confidence intervals in statistical analysis?
~ Confidence intervals show the range where a population value likely lies, with a given confidence level (like 95%).
Significance:
Reflects the uncertainty in estimates
Helps in decision-making without needing the full population
Tells how reliable a sample estimate is

19. What is the relationship between a Z-score and a confidence interval?
~ A Z-score helps calculate a confidence interval when the population standard deviation is known.
Relationship:
Confidence intervals use Z-scores to set the range around the sample mean.

20. How are Z-scores used to compare different distributions?
~ Z-scores standardize values from different distributions, making them comparable.
How:
Convert values to a common scale (mean = 0, standard deviation = 1)
This lets you compare how far each value is from its own mean

21. What are the assumptions for applying the Central Limit Theorem?
~Assumptions of the Central Limit Theorem (CLT):
Random Sampling – Data must be randomly selected.
Independence – Each sample should be independent of others.
Sample Size – Should be large enough (usually n ≥ 30).
Finite Variance – Population must have a finite standard deviation.

22. What is the concept of expected value in a probability distribution?
~ The expected value is the average outcome you’d expect if an experiment is repeated many times.

23. How does a probability distribution relate to the expected outcome of a random variable?
~ A probability distribution shows all possible values of a random variable and their probabilities.
The expected outcome (expected value) is calculated from this distribution by weighting each value by its probability.

```python
#2 Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the discrete uniform distribution (like a fair die)
low, high = 1, 7  # randint is exclusive of upper bound, so use 7 for 1–6

# Create a discrete uniform distribution object
dist = randint(low, high)

# Generate values and PMF
x = np.arange(low, high)
pmf_values = dist.pmf(x)

# Plot the PMF
plt.stem(x, pmf_values, basefmt=" ")
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.xticks(x)
plt.grid(True)
plt.show()
```

```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

from scipy.stats import bernoulli

def bernoulli_pdf(p, x):
    """
    Calculate the PDF of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (0 ≤ p ≤ 1)
    x (int): Outcome (0 or 1)

    Returns:
    float: Probability of observing x
    """
    if x not in [0, 1]:
        return 0  # Invalid input for Bernoulli
    return bernoulli.pmf(x, p)

# Example usage
p = 0.7  # Probability of success
print("P(X=0):", bernoulli_pdf(p, 0))  # Probability of failure
print("P(X=1):", bernoulli_pdf(p, 1))  # Probability of success
```

```python
#9 Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot their histogram.

    Parameters:
    data (list or np.array): Input dataset
    """
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=30, color='mediumseagreen', edgecolor='black', alpha=0.7)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage with a normal dataset
sample_data = np.random.normal(50, 10, 1000)
z_scores = calculate_and_plot_z_scores(sample_data)
```

```python
#10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution..

import numpy as np
import matplotlib.pyplot as plt

def simulate_clt(sample_size=30, num_samples=1000):
    """
    Demonstrate CLT using means of samples from an exponential distribution.

    Parameters:
    sample_size (int): Number of values in each sample
    num_samples (int): Number of samples to draw
    """
    sample_means = []

    for _ in range(num_samples):
        sample = np.random.exponential(scale=2.0, size=sample_size)  # Non-normal distribution
        sample_mean = np.mean(sample)
        sample_means.append(sample_mean)

    # Plot histogram of sample means
    plt.hist(sample_means, bins=30, density=True, color='lightblue', edgecolor='black')
    plt.title(f"CLT Demonstration\nSample Size = {sample_size}, Samples = {num_samples}")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.grid(True)
    plt.show()

# Run the simulation
simulate_clt(sample_size=30, num_samples=1000)
```

```python
#14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculate Z-score for a given data point.

    Parameters:
    x (float): Data point
    mean (float): Mean of the distribution
    std (float): Standard deviation of the distribution

    Returns:
    float: Z-score
    """
    return (x - mean) / std

def plot_z_on_normal(z):
    """
    Plot the standard normal distribution and mark the Z-score.

    Parameters:
    z (float): Z-score to highlight on the curve
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.plot(x, y, label="Standard Normal PDF", color='blue')
    plt.fill_between(x, 0, y, where=(x <= z), color='lightcoral', alpha=0.5, label=f'Area ≤ Z={z:.2f}')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    plt.title("Standard Normal Distribution with Z-score")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.legend()
    plt.show()

# Example input
data_point = 75
mean = 70
std_dev = 5

# Calculate Z-score and plot
z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point}:", round(z_score, 2))
plot_z_on_normal(z_score)
```

```python
#15 Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def z_test(sample_data, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    sample_data (list or np.array): Sample data
    population_mean (float): Hypothesized population mean (H0)
    population_std (float): Known population standard deviation
    alpha (float): Significance level (default=0.05)

    Returns:
    None
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)

    # Calculate Z-statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Two-tailed critical Z-value
    z_critical = norm.ppf(1 - alpha/2)

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z:.2f}")
    print(f"Z-critical (±{z_critical:.2f}) at α = {alpha}")

    # Hypothesis decision
    if abs(z) > z_critical:
        print("🔴 Reject the null hypothesis (significant difference).")
    else:
        print("🟢 Fail to reject the null hypothesis (no significant difference).")

# Example sample dataset
sample_data = [72, 68, 70, 74, 69, 71, 70, 73, 68, 72]

# Known population parameters
population_mean = 70
population_std = 3

# Run Z-test
z_test(sample_data, population_mean, population_std)
```

```python
#16 Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a dataset.

    Parameters:
    data (list or np.array): Sample data
    confidence (float): Confidence level (default = 0.95)

    Returns:
    (lower_bound, upper_bound): Confidence interval
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # sample standard deviation
    n = len(data)
    z = norm.ppf(1 - (1 - confidence) / 2)  # Z-score for given confidence
    margin_of_error = z * (std / np.sqrt(n))

    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print("Interpretation: We are", int(confidence*100), "% confident that the true population mean lies within this range.")

    return (lower, upper)

# Example sample dataset
sample_data = [72, 68, 70, 74, 69, 71, 70, 73, 68, 72]

# Calculate and interpret the confidence interval
confidence_interval(sample_data, confidence=0.95)
```

```python
#17 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy.stats import norm

def generate_data_and_ci(mean=50, std=10, size=100, confidence=0.95):
    # Step 1: Generate data from a normal distribution
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Step 2: Calculate sample mean and standard deviation
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # sample standard deviation
    z = norm.ppf(1 - (1 - confidence) / 2)  # z-score for confidence level
    margin_error = z * (sample_std / np.sqrt(size))

    # Step 3: Compute the confidence interval
    lower = sample_mean - margin_error
    upper = sample_mean + margin_error

    # Step 4: Print results and interpretation
    print(f"Generated Sample Mean: {sample_mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower:.2f}, {upper:.2f})")
    print(f"Interpretation: We are {int(confidence*100)}% confident the true population mean lies within this interval.")

# Run the function
generate_data_and_ci(mean=50, std=10, size=100, confidence=0.95)
```

```python
#18 Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean=0, std=1):
    """
    Calculate and plot the PDF of a normal distribution.

    Parameters:
    mean (float): Mean of the distribution
    std (float): Standard deviation of the distribution
    """
    # Generate x values
    x = np.linspace(mean - 4*std, mean + 4*std, 1000)

    # Calculate PDF values
    pdf = norm.pdf(x, loc=mean, scale=std)

    # Plot
    plt.plot(x, pdf, color='blue', label=f'N({mean}, {std}²)')
    plt.title("Normal Distribution PDF")
    plt.xlabel("x")
    plt.ylabel("Probability Density")
    plt.grid(True)
    plt.legend()
    plt.show()

# Run the function for N(0,1)
plot_normal_pdf(mean=0, std=1)
```

```python
#22 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np
import matplotlib.pyplot as plt

# Generate dataset from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=100, scale=15, size=1000)  # Mean=100, SD=15

# Calculate range
data_range = np.max(data) - np.min(data)

# Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

print(f"Range: {data_range:.2f}")
print(f"IQR (Interquartile Range): {iqr:.2f}")

# Visualize with a boxplot
plt.boxplot(data, vert=False)
plt.title('Boxplot of Normally Distributed Data')
plt.xlabel('Value')
plt.grid(True)
plt.show()
```

```python
#23 Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

# Generate dataset from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=100, scale=20, size=1000)

# Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_normalized_data = (data - mean) / std

# Plot original and normalized data
plt.figure(figsize=(12, 5))

# Original data histogram
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Z-normalized data histogram
plt.subplot(1, 2, 2)
plt.hist(z_normalized_data, bins=30, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Print mean and std of normalized data
print(f"Mean after normalization: {np.mean(z_normalized_data):.2f}")
print(f"Standard deviation after normalization: {np.std(z_normalized_data):.2f}")
```

```python
#24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_distribution_stats(mean=0, std=1, size=1000):
    # Generate data from normal distribution
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skew_val = skew(data)
    kurt_val = kurtosis(data, fisher=False)  # fisher=False => regular kurtosis (not excess)

    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis: {kurt_val:.4f}")
    return skew_val, kurt_val

# Example usage
analyze_distribution_stats(mean=100, std=15, size=1000)
```
